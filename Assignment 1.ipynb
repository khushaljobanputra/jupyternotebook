### Pandas and exploring/representing the data using MatPlotLib

```python
boston = load_boston()
boston.keys()
```

```python
#provides us with details about the dataset i'm using for this example
print (boston.DESCR)
```

```python
bos = pd.DataFrame(boston.data)
bos.columns = boston.feature_names
bos['PRICE'] = boston.target
bos.head()
```

```python
#plotting a individual feature of the dataset
bos['CRIM'].plot()
plt.show()

#plotting histogram for all the features of the dataset
bos.hist(figsize=(16,12))
plt.show()
```

### Exploring SciKit Learn

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(bos, test_size=0.2, random_state=42)

print(train_set.describe())
print()
print("Training Percentages:")
print()
print(train_set.count() / bos.count())
```

```python
train_set.plot(kind="scatter", x="CRIM", y="AGE", alpha=0.3)
plt.show()
```

```python
train_set.plot(kind="scatter", x="DIS", y="RAD", alpha=0.3, s=bos["CRIM"], label="CRIM", 
               figsize=(10,8), c="AGE", cmap=plt.get_cmap("jet"), colorbar=True, sharex=False)
plt.legend()
plt.show()
```
